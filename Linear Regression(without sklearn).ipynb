## reading data into a datframe

```python
data = pd.read_csv('/home/varun/Downloads/data.csv')
```

```python
X = data.iloc[:,0]
Y = data.iloc[:,1]
```

```python
print(X.max())
print(X.min())
```

```python
newX = (X - X.min())/(X.max()-X.min())
newY = (Y - Y.min())/(Y.max()-Y.min())
```

```python
num = X.shape[0]
```

```python
num_of_rows = int(num*0.8)
```

```python
x_train = newX.iloc[:num_of_rows] 
x_test = newX.iloc[num_of_rows:]
y_train = newY.iloc[:num_of_rows]
y_test = newY.iloc[num_of_rows:]
```

## Initializing parameters


```python
m = 0
c = 0
```

```python
"""
Learning rate

"""
alpha = 0.001
```

```python
n = num_of_rows
```

```python
itr = 100000
for i in range(itr):
    y_pred = m*x_train + c
    cost = (1/2*n)*sum((y_pred-y_train)**2)
    der_m = (1/n)* sum(x_train*(y_pred-y_train))
    der_c = (1/n)* sum(y_pred-y_train)
    m = m - alpha * der_m
    c = c - alpha * der_c
    sys.stdout.write("\rCost: %i  m: %f  c: %f" % (cost,m,c))
    sys.stdout.flush()
```
